# Dichotomie
## Exercice 1 : liste d'amis
Nicolas organise une fête, et demande à ses amis s'ils viendront. Dès qu'un ami lui répond favorablement, il l'ajoute dans liste_amis.  
Compléter le code ci-dessous afin de pouvoir déterminer si Vincent, Romain et Valérie ont décidé de venir (bien respecter les majuscules, minuscules et accents).
<div class="alert alert-danger"> 
Vous devez absolument réaliser une recherche dichotomique et pas une recherche naïve.
    </div>

````python
ma_liste = ['Ariel', 'Benjamin','Amir','Melvina','Hector','Zoé','Yasmine','Ursula',\
              'Andréa','Marie-Claire','Marc-Aurèle','Jim','Paula','Pietro','Xavier','Eléonore',\
              'Arthur','Firmin','Valérie','Alice','Bob','Firdaous','Olga','Shinzo','Li','Mathéo','Rachel',\
              'Philippine','Tao']
````

````python
def appartient_dichotomique(ma_liste, cible):
    ma_liste.sort()
    indice_debut = 0
    indice_fin = len(ma_liste) - 1

    while indice_debut <= indice_fin:
        indice_centre = (indice_debut + indice_fin) // 2
        valeur_centrale = ma_liste[indice_centre]

        if valeur_centrale == cible:
            return True
        elif valeur_centrale < cible:
            indice_debut = indice_centre + 1
        else:
            indice_fin = indice_centre - 1

    return False
````

````python
# Test
assert not appartient_dichotomique(ma_liste, "Vincent")
assert not appartient_dichotomique(ma_liste, "Romain")
assert appartient_dichotomique(ma_liste, "Valérie")
````

## Exercice 2 : panne système
Une sonde interroge à intervalles réguliers l'état de fonctionnement d'un système électronique. Celui-ci peut être en marche ou en panne.

La sonde est programmée pour enregistrer les résultats de ses requêtes dans un *log*. Il s'agit d'un tableau de booléens (une liste Python) dans lequel les valeurs True précèdent les False. La valeur True indique que le système est en marche, False qu'il est en panne.

Une panne nécessite une intervention humaine et ne peut donc pas disparaître seule : elle persiste jusqu'à la fin de l'enregistrement. 
Exemple de *log* :
``` Python
#         0     1      2      3      4
log = [True, True, False, False, False]

```
Lors d'une vérification on constate que le système est en panne : le *log* contient au moins une valeur False en dernière position. On se demande à quel moment a débuté cette panne.

Dans l'exemple précédent, le premier False est à l'indice 2 : la panne a débuté à l'instant 2.



Écrire la fonction `indice_panne` qui prend en paramètre le tableau de booléens `log` et renvoie l'instant du démarrage de la panne. 
<div class="alert alert-danger"> 
    Il est interdit de faire une recherche naïve c'est-à-dire de passer en revu les valeurs une à une. 
</div>
On garantit que le `log` n'est pas vide et que, au moment de la vérification, le système est en panne (la dernière valeur du tableau est `False`).

````python
def indice_panne(log):

    indice_debut = 0
    indice_fin = len(log) - 1

    while indice_debut <= indice_fin:
        indice_centre = (indice_debut + indice_fin) // 2
        valeur_centrale = log[indice_centre]

        if valeur_centrale == False and log[indice_centre -1] == True:
            return indice_centre
        elif valeur_centrale == False:
            indice_debut = indice_centre + 1
        else:
            indice_fin = indice_centre - 1

    return None
````

````python
indice_panne([True, False])
````

````python
indice_panne([True, True, False, False, False])
````

````python
from random import randint

N = 10**5
indice = randint(0, N)
print(indice)
mon_log = [True if i < indice else False for i in range(N)]
print(mon_log)
solution = indice_panne(mon_log)
solution
````

## Exercice 4 : complexité de l'algorithme de recherche dans une matrice carrée
On donne l'algorithme de recherche d'une valeur `cible` dans une matrice `mat` de longueur et de largeur `n` :


début  
&nbsp;&nbsp;solution ← faux  
&nbsp;&nbsp;pour i allant de 0 à n-1  
&nbsp;&nbsp;&nbsp;&nbsp; pour j allant de 0 à n-1  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; si mat[i][j] = cible  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; alors solution ← vrai  
&nbsp;&nbsp;&nbsp;&nbsp; fin pour  
&nbsp;&nbsp;fin pour  
fin  

1. Déterminer le nombre d'opérations élémentaires nécessares pour une matrice `mat`de longueur `n`.
2. En déduire la complexité de cet algorithme.

3. Le nombre d'opérations élémentaires nécessaires pour cet algorithme est proportionnel à la longueur du tableau `n`, car chaque itération de la boucle effectue une opération élémentaire.

4. La complexité de cet algorithme est linéaire, notée O(n), car le temps d'exécution croît linéairement avec la taille de l'entrée (la longueur du tableau).

1) 1+n*n*(1+1) = 1+2n^2
2) La complexité est de l'ordre de O(n^2). C'est une complexité quadratique.

````python
from timeit import timeit

def recherche(matrice: list, x: int) -> int :
    """
    :param matrice: une tableau 2D
    :param x: de type int
    :returns: bool : True si x est dans tableau, False sinon

    >>> recherche([1, 3, 4, 9], 9)
    True
    >>> maxi([1, 3, 4, 9], 5)
    False
    """
    solution = False
    for i in range(len(matrice)) :
        for j in range(len(matrice[i])) :
            if matrice[i][j] == x :
                solution = True
    return solution

tailles = [10, 30, 90, 270, 810]

# tref est le temps de calcul pour une liste de taille 100
liste_ref = [[i+j for j in range(10)] for i in range(10)]
tref = timeit("recherche(liste_ref, -1)", number = 1000, globals = globals())
print("n = 10 -> tref = ",round(tref, 6))

for n in tailles :
    print("n =", n, end='')
    tab = [[i+j for j in range(n) ] for i in range(n)]
    # Calcul du temps pour des listes triées de tailles n
    t = timeit("recherche(tab, -1)", number = 1000, globals = globals())

    print('\t-> temps = ',round(t, 6), '\t x', round(t/tref, 2) )
    tref = t
````
