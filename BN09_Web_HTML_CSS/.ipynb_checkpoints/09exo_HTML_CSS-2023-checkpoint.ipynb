## étape 2 : réduire la taille de l'image de Tim Berners Lee
Ajoutant un attribut de taille en pixel dans la balise image de Tim Berners Lee :  
````html
<img src="mon_image.JPG" height="352" width="470" >
````

**Astuce :** Si vous ne mentionnez que la largeur (ou la hauteur) alors l'autre dimension sera calculée par le navigateur et le ratio sera conservé.

**Vous auriez préféré une taille d'image en pourcentage ?**  
Bien que les pourcentages soient compris par les navigateurs `<img src="mon_image.jpg" height="200%">`  
ils ne sont plus autorisés par le W3C, une solution existe avec le CSS :

````css
img {
	max-width: 100%;
	height: auto;
}
````

## étape 6 : ajouter une barre de navigation
Ajouter une barre de navigation :


Le HTML 5 possède une balise `nav` à cet effet. Mais cette balise ne produit pas d'effet, elle est équivalente à une balise `div`. Son seul intérêt est d'informer que le contenu de cette balise `nav` correspond à un menu.

**Facultatif :**
Le plus souvent, une navbar est une liste `ul` dont les items `li` sont disposés :
* horizontalement sur grand écran
* ou verticalement sur un smartphone (grâce aux Media Queries).

Page HTML :
````html
<nav id="menu">
   <ul>
      <li><a href="page1.html">page 1</a></li>
      <li class="ici">page 2</li>       
      <li><a href="page3.html">page 3</a></li>
   </ul>
</nav>
````

Feuille de style CSS :
````css
nav ul {list-style-type: none; /* une liste sans puce */  }

nav ul li  {
  display: inline-block ;  /* disposé côte à côte (= aligné) */
  vertical-align: bottom;  /* sur une ligne en bas */
  height: 30px;    line-height: 30px;  /* hauteur = 30 = interligne */
  padding: 0px 10px 0px 10px; /* marge interne pour le texte contenu */
  background-color: rgb(116,97,205);    /* bleu assez clair */
}

nav ul li.ici {  /* onglet « page en cours » */
  height: 40px;  line-height: 40px; /* plus haut : hauteur = 40 = interligne */
  background-color: rgb(255,123,140); /* onglet rouge */
  font-size: 1.3em;  text-decoration: underline;  
}	
		
@media all and (max-width: 800px){
  nav { display: inline-block ; /* FACULTATIF : un petit bloc */  }
 
  nav ul li { display: block ; /* en lignes de même grandeur */   }
 
}
````


**Comment réaliser un lien interne à la page web ?**
Par exemple pour réaliser un lien qui pointe vers un titre `h2` :
* il faut ajouter un identifiant `id` dans cette balise (en attribut) :  
````html
<h2 id="CPU_avec_pipeline">Exemple de CPU avec pipeline</h2>
````
* écrire le lien interne `a` pointant vers cet identifiant grâce symbole `#` :
````html
S O M M A I R E :
<a href="#CPU_avec_pipeline">Exemple de CPU avec pipeline</a> <br>
<a href="#CPU_sans_pipeline">Un CPU sans pipeline</a>
````

**A retenir :**
> Une `class` peut être utilisée plusieurs fois dans le document,  
> et dans le CSS : `.` désigne une **classe**.

> En revanche, un `id` doit être **unique** dans la page HTML,  
> et dans le HTML et CSS : `#` désigne un **identifiant**.

## Facultatif : sommaire verticale


Dans le fichier HTML :
````html
<main>
  <nav id="sommaire">
    <h2>S O M M A I R E :</h2>
    <ol>
      <li><a href="#mon_titre1">titre 1</a></li>
      <li><a href="#mon_titre2">titre 2</a></li>
    </ol>
  </nav>

  <section>.................... </section>

  <aside>.......................</aside>
<main>

<footer> ..... </footer>
````

Dans le fichier CSS :

````css
html, body { height: 100%;  margin: 0; padding : 0;}

main {
    text-align: center;
    padding-bottom: 20px; /* hauteur du footer */
}

section {width: 300px;}

aside {width: 150px;}

footer {
    position: fixed;
    bottom: 0;
    left: 0;
    height: 20px; /* hauteur du footer */
    width: 100%;
    background-color: lightcyan;
}

nav#sommaire, section, aside {
  display: inline-block;
  margin: 20px 0px 20px 20px; /* top right bottom left */
  background-color: rgb(170,185,242); /* bleu */
  vertical-align: top;
}
````
