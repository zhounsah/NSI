# Le CSS
Il est possible de modifier l'affichage d'une page web en ajoutant des propriétés CSS sur des balises HTML.  
CSS (Cascading Style Sheets) = feuilles de style en cascade.

## Un peu d'histoire ..
Il y a eu plusieurs tentatives de projets sur le style.  
Mais c'est en 1997, que démarre véritablement le CSS, il comporte une cinquantaine de propriétés.  
Puis le niveau 2 du CSS s'enrichit de 70 nouvelles propriétés. Puis le niveau 3 (qui n'est pas encore terminé).

Le CSS est développé par « niveaux » où chaque nouvelle couche contient les niveaux précédents.  
La rétro-compatibilité est donc assurée.

Dans le CSS 3, certains modules ne sont toujours pas officialisés .. mais des versions de travail sont néanmoins implémentées par les navigateurs.

## Style en ligne (dans chaque balise HTML)
Exemple :
```html
<h6 style="color: purple; text-shadow: 1px 1px darkorange; font-variant: small-caps;">
    Un 1er titre h6 avec un style en ligne
</h6>
<h6>Un 2e titre h6 (sans style .. lui)</h6>
```


**Inconvénient** : Ici on déclare les propriétés CSS et leurs valeurs directement dans la balise HTML à "styler" par un attribut `style`.  
Et **il faut le faire dans chaque balise !!!!** Donc c’est très répétitif et le HTML se retrouve noyé !   
==> La méthode du style en ligne est **déconseillée** !

## Style sur un sélecteur

Le style est déclaré par l'intermédiaire d'un sélecteur qui pointe vers TOUS les éléments HTML visés.  
Ces propriétés sont placées :


>* soit **dans une balise `<style>`** à mettre dans l'entête `<head>` de la page HTML

>* soit **dans un fichier `.css`** et ajouter un lien de *relationship* `rel`  dans l'entête `<head>` de la page HTML :

```html
<link rel="stylesheet" href="style.css">
```

Exemple de fichier CSS ou contenu de balise `<style>` :
````css
body {color:red;}

h6 {
    color: purple;
    text-shadow: 1px 1px darkorange;
    font-variant: small-caps;
}

p {font-weight:bold;}
````

**L’avantage** : Un fichier CSS permet d'avoir le même CSS sur toutes les pages web sans recopier !  

NB : Si on veut modifier la mise en page d’UNE page, il suffit de rajouter un 2ème lien vers un autre fichier css qui surchargera le 1er !  
C’est le principe des « feuilles de style en cascade ».

## Style en cascade
Il est tout à fait possible d'utiliser toutes ces formes dans une même page web : 
* style en ligne (incorporé à une balise)
* style en interne (balise `<style>` dans `<head>`)
* style externe (fichier CSS).

En cas de recouvrement ou contradiction, certaines instructions seront prioritaires !
> **Plus l’instruction CSS a une cible précise .. plus elle est prioritaire !**  
ex : `<p>` est prioritaire par rapport à `<body>` car `<p>` est à l'intérieur de `<body>`.

> En cas de succession de sélecteurs : c’est **la dernière propriété qui est prioritaire**.  
ex : `a {color: red;}` puis `a {color: navy;}` : c’est la dernière propriété 'navy' qui est prioritaire.

> En cas de CSS en ligne et en interne et en externe : 
* c'est le **style en ligne qui est prioritaire** !  
* En 2e : le style en **interne : la balise `<style>`**.  
* En dernier : le **fichier CSS**. 

## Quelques propriétés CSS
### texte
* `font-style: italic;`  pour mettre en italique.  
`font-style: normal;`  pour ne **pas** mettre en italique.


* `font-weight: bold;`  pour mettre en gras.  
`font-weight: normal;`  pour ne **pas** mettre en gras.


* `text-decoration: underline;` pour souligner  
`text-decoration: none;` pour ne **pas souligner**


* `font-size: 1.6 em;` pour mettre une taille d'écriture de texte à 160% de la taille normale.


* `font-family: Impact, "Arial Black", Verdana, Serif;` propose plusieurs polices de caractères (car le client ne les possède peut-être pas).


* `font-variant: small-caps;` pour mettre un texte en petites capitales -OU- grandes capitales : `text-transform: uppercase;`

### colorier
**Liste des couleurs connues** : [https://en.wikipedia.org/wiki/Web_colors](https://en.wikipedia.org/wiki/Web_colors)

* `color: navy;` pour un **texte** de couleur bleu marine.  
`color: rgb(25, 0, 118);` pour une couleur personnalisée  
`color: rgba(25, 0, 118, 0.5);` pour ajouter de la transparence avec le canal Alpha


* `background-color: yellow;` pour une couleur **de fond** (tel un stabilo).

**pour faire télécharger une police de caractère** :  
Mettre le fichier `3Dumb.ttf` dans un dossier `polices` :
````css
@font-face { /* dans un dossier 'polices' */
  font-family: '3Dumb';
  src: url('polices/3Dumb.ttf'); 
}
p {font-family: '3Dumb';}
````

### block
* `background-color: green;` pour une couleur de fond sur tout le bloc.


* `margin: 10px;` pour une **marge extérieure** de 10px tout autour du bloc  
`margin: 10px 3px 5px 3px;`  dans l’ordre : top | left | bottom | right


* `padding: 5px;` pour une **marge interne** (sur tous les côtés)


* `border-style: solid;` pour encadrer et `border-width: 5px;`pour l'épaisseur.


* `table {border-collapse: collapse;}` pour coller les cellules entre elles dans un tableau. 
* `td, th {border: 1px solid black;}`  pour encadrer chaque cellule. Par défaut, les bordures ne sont pas tracées.

### Les changements d’état
* `a:hover {letter-spacing: 2px;}` change l'espacement des lettres **lorsque la souris survole le lien `<a>`**

Les pseudo-classes ( :hover :active … ) fonctionnent aussi sur d’autres balises : `<p>`, `<li>`, … ainsi que les `<button>` ou champs de formulaires `<select> <input>`.
* `:hover`    : au survol de la souris.
* `:focus`    : lorsqu’un élément sélectionné (un clic dans un champ de formulaire).
* `:disabled`  : lorsque l'élément est désactivé (un bouton par exemple).
* `:active`  : au moment du clic (c’est un temps très très court)

## Sommaire = ancres internes à la page


**Comment réaliser un lien interne à la page web ?**
Par exemple, pour réaliser un lien qui pointe vers un titre `h2` :
* il faut ajouter un identifiant `id` dans cette balise (en attribut) :  
````html
<h2 id="CPU_avec_pipeline">CPU avec pipeline</h2>
````
* écrire le lien interne `a` pointant vers cet identifiant grâce symbole `#` :
````html
S O M M A I R E :
<a href="#CPU_avec_pipeline">CPU avec pipeline</a> <br>
<a href="#CPU_sans_pipeline">CPU sans pipeline</a>
````

**A retenir :**
> Une `class` peut être utilisée plusieurs fois dans le document,  
> et dans le CSS : `.` désigne une **classe**.

> En revanche, un `id` doit être **unique** dans la page HTML,  
> et dans le HTML et CSS : `#` désigne un **identifiant**.

**Dans le CSS, comment différencier les 2 menus de navigation ?**  
* En mettant un identifiant à chacun dans le HTML.<br>Par exemple : `<nav id="menu">` et `<nav id="sommaire">` 


* En faisant référence à cet identifiant dans le CSS : `#menu` (ou `nav#menu`) et `#sommaire` (ou `nav#sommaire`).

```css
/* 1er menu : #menu */
nav#menu ul {list-style-type: none;  }
nav#menu ul li  { ... }

/* 2e menu : #sommaire */
nav#somaire ul { .... }
```

# Chemin vers un fichier
## Chemin relatif ou absolu
Comment naviguer dans l'arborescence ? .. Le principe est simple :

* `../moi.png` Remonte dans le **dossier parent** et prend l'image "moi.png" (c'est un chemin relatif).


* `/moi.png` Remonte à la **racine du site web** et prend l'image "moi.png" (c'est un chemin absolu car son référentiel n'est pas la page web, mais c'est un répertoire fixe).


* `moi.png` ou `./moi.png` :  Prend l'image "moi.png" dans le **dossier courant** (donc juste à côté de la page web actuelle)(c'est un chemin relatif).


* `image/moi.png` Prend l'image "moi.png" dans le **sous-répertoire** "image" (c'est un chemin relatif).


>Conclusion : Un chemin absolu commence par le caractère `/` qui désigne *la racine* du site.

## URL relative ou absolue
Une URL est un chemin sur le web, comme une adresse de site web. Ex : `https://nsirennes.fr`
* une **URL absolue** indique le chemin complet (donc elle toujours par « http:// ») ex : `<a href='https://nsirennes.fr/wp/uploads/menu.svg'>jolie image</a>`

* une **URL relative** indique le chemin :
    * à l'aide d'un chemin absolu, donc depuis la racine du site web. ex : `<a href='/wp/uploads/menu.svg'>jolie image</a>` )
    * ou d'un chemin relatif, donc depuis la page web qu'on écrit. ex : `<a href='uploads/menu.svg'>jolie image</a>`
