# Les variables et leurs types
Le contenu de chaque variable est stocké en mémoire RAM, dans une « case numérotée » (appelée adresse mémoire).  
NB : La RAM (Random Access Memory) est une mémoire volatile (son contenu est perdu lorsqu'on éteint l'ordi).


<span class="enonce">1/ Quel est le type de chacune des variables ?<br>2/ Exécuter le programme pour vérifier.<br>3/ Faire afficher le type de la variable b.</span>

    1) La variable nom est une variable composéee de texte tandis que les autres variables sauf b sont des nombres entiers naturels. La variable b est booléen.

```python
nom = "Kevin"
age = 16
x = 23
y = x
b = False
# on affiche le type des variables nom, age et x.
print("nom :",type(nom),"| age : ",type(age),"x :",type(x), "b :", type(b))

# afficher le type de la variable b
```

Une variable, c'est un nom, une valeur et un type.  
Le type de la variable est déterminé lors de l'affectation.  

Les types de base sont :   
* `int` : nombre entier (en anglais : integer).
* `float` : nombre à virgule flottante. /!\ La virgule est notée par un `.`
* `str` : texte (chaîne de caractères, string).
* `bool` : booléens qui sont `True` et `False` (noter la majuscule en Python).

Et, il existe beaucoup d'autres types de variable ...

`type()` : fonction qui donne le type de la variable.

```python
## Quelques exemples :
a = 1/3
print("a =",a,"est de type",type(a))

b = 6/2  # division décimale
print("b =",b,"est de type",type(b))

c = 6//2  # division entière
print("c =",c,"est de type",type(c))

d = ""  # chaine vide
print("d =",d,"est de type",type(d))

e = True
print("e =",e,"est de type",type(e))

f = (5 == 5.0)
print("f =",f," et son type est",type(f))
```

D'ailleurs, l'instruction `a = 1` peut aussi être écrite en précisant le type : `a:int = 1`

## print()
La fonction `print()` peut afficher plusieurs éléments séparés par des virgules.  
(Ce sont des arguments pour la fonction print).  

```python
# print avec 1 seul argument
print('Hello World !')

# print avec 3 arguments
votre_nom = "Donald"
print('Bonjour', votre_nom, 'et bienvenue en NSI!')

# OU BIEN : on utilise une f-string (formatted string)
print(f'Bonjour {votre_nom} et bienvenue en NSI!')
```

Par défaut, `print()` est configuré ainsi :
* le séparateur est un espace `sep = " "` , mais on peut le changer. *Voir ci-dessous*.  
* le caractère de fin est un 'retour à la ligne' `end = "\n"`, idem, on peut le changer.

```python
print('Bonjour', 'à', 'la Terre entière','.', sep="***")
```

```python
print('Hello', end = "")  # le caractère de fin est le vide
print('Brian Smith', end = "")
print('Alan Grant', end = "")
print('.')
```

## input()
La fonction `input` provoque une interruption du programme .. invitant l'utilisateur à saisir des caractères.

**/!\ TOUTES LES ENTREES `input` SONT DE TYPE TEXTE `str`.  
Donc très très souvent, on caste l'input en entier ou flottant**.  
On peut aussi caster en booléen, si besoin. 

```python
# votre_nom est automatiquement un string
votre_nom = input('Quel nom ? ')

# la saisie de l'age donne un texte, puis on caste ce texte en entier
votre_age = int(input('Quel age ? '))
print(votre_nom,", l'an prochain, vous aurez",votre_age + 1,"ans.") 
```

NB : Si le texte saisi par l'utisateur n'est pas un entier, alors on a une **erreur Python**.

|---|---|---|

| `bool()`  | transforme une variable en booléen.  | `vrai = bool(1)` donne True  |

**/!\ Attention, le transtypage n'est pas toujours possible !!!** Ci-dessous, un exemple :

```python
int("2a")
```

## Les 2 divisions
> Il y a 2 divisions : 
> * **la division décimale `/`** : le résultat est un flottant (ex : 6/2 donne 3.0)
> * **la division entière `//`**  



```python
### Division décimale :  /
a = 7/4
# ===> Afficher le résultat de la division décimale 7/4
print (a)



### Division entière : //
b = 7//4
# ===> Afficher le résultat de la division entière 7//4
print (b)
```

NB : Ces 2 divisions fonctionnent aussi avec les flottants (et le résultat obtenu est de type flottant).

```python
# Diviser 156.78 par 2.93 (avec les 2 divisions)

aa = 156.78/2.93
ab = 156.78//2.93

print("Division décimale :", (aa),"\nDivision entière :",(ab))

# Diviser 6 par 2 (avec les 2 divisions)

ba = 6/2
bb = 6//2
```

## Davantage sur les sous-chaines
Soit `chaine` est une chaîne de caractères :
> * **` chaine[i] `** : l'élément d'index i (sachant que le 1er index est 0).
> * **` chaine[-1] `** : le dernier élément. De même, chaine[-2] est l'avant-dernier élément ...

> * **` chaine[i:j] `** : renvoie une partie de la séquence de l'index i (inclus) à l'**index j exclu**. (appelé slicing)

```python
texte = "Les 101 dalmatiens"
a = texte[2]
print(f'le caractère d\'index 2 est : {a}.')

b = texte[2:6]
print(f'le slicing est  : {b}.')
```

> * **` chaine.index(x) `** : renvoie l'index de la 1ère apparition de x dans la séquence. Et renvoie `ValueError` si x en est absent.  
**/!\ la numérotation commence à 0 : donc l'index du 1er caractère est 0. Le 2e caractère a pour index 1 etc ...**
> * **` chaine.count(x) `** : renvoie l'effectif de x dans la séquence. Et renvoie `0` si x en est absent.  

On `index` et `count` sont des fonctions qui s'exécutent sur l'objet `chaine` grâce au point `.` : on dit que ce sont des **méthodes**.

```python
refrain = "Darla dirladada"

# la position de la 1ère occurence :
position = refrain.index("da")

# le nb d'occurences
nb = refrain.count("da")
print(f"position = {position} et nb = {nb}")
```
