## Exercice 12 : QCM fonction et boucle `while`

Soit n un entier naturel.  
Sa factorielle est le produit des nombres entiers strictement positifs qui sont plus petits ou égaux à n.  
Par exemple : la factorielle de 4 vaut  4\*3\*2\*1 donc 24.

Parmi ces fonctions, une seule retourne bien la factorielle de n. Lequel ? .....

**Auto-correction :** Ajouter l'appel console `factorielle(4)` et lancer le programme.

````python
# Réponse A :
def factorielle(n):
    i = 0
    fact = 1
    while i <= n:
        fact = fact * i
        i = i + 1
    return fact
````

## Exercice 13 : 
Ecrire un programme qui utilise une boucle <code>for</code> pour tracer un carré de côté 100 pixels, avec un contour bleu et colorié en vert.

````python
import turtle

carré = turtle.Turtle()
carré.color("blue", "green")
carré.penup()
carré.goto(-50, -50)  # Déplacer le carré pour qu'il soit centré

# Dessiner le carré
carré.begin_fill()
carré.pendown()
for _ in range(4):
    carré.forward(100)  # Longueur du côté
    carré.left(90)  # Angle droit
carré.end_fill()

# Fermer la fenêtre en cliquant dessus
exitonclick()
````
