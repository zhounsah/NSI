Ecrire une fonction `perimetre_carre` qui avec le paramètre `côté` renvoie le périmètre du carré.  
Compléter la docstring.

````python
def perimetre_carre(cote):
    """
    Calcule le périmètre d'un carré à partir de la longueur de son côté.

    PARAM
    -----
    cote (float): Longueur du côté du carré.

    RETURN
    -------
    float: Périmètre du carré.

    EXAMPLES
    --------
    >>> perimetre_carre(3)
    12.0
    >>> perimetre_carre(4.7)
    18.8
    >>> perimetre_carre(3/8)
    1.5
    """
    return 4 * cote
````
