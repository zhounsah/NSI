### Phase 4 : par somme de puissances de 2
Pour bien communiquer avec des machines il est donc nécessaire de savoir convertir un nombre décimal en un nombre binaire.

Une méthode de conversion consiste à :
> **décomposer le nombre décimal en une somme de puissances de deux : 2<sup>7</sup>, 2<sup>6</sup>,  ... , 2<sup>0</sup>**.

Il s’agit tout simplement de la méthode des cartes binaires, … qu’on peut étendre à 8 bits (8 cartes).  

Effectivement :  chaque carte représente une puissance de 2.  
- le coefficient est **``1``** si elle est tournée du côté visible (recto) ;
- le coefficient est **``0``** si la face est invisible (côté verso).

> **Le nombre est alors la somme des puissances de 2 (c’est-à-dire les cartes visibles).**

Par exemple, pour la conversion : 91 (décimal) = `01011011`(binaire)

## A retenir sur le binaire
- **Codage** : Opération consistant à représenter des informations à l'aide d'un code.
- **Codage binaire** : Le code binaire utilise exclusivement les symboles `0` et `1` (systèmes logiques).
- **Bit** : C'est le chiffre élémentaire de la numérotation binaire.
- **Mot** : Groupe de "n" bits. **Un mot 8 bits s'appelle un _octet_.**
- **Poids** : Coefficient attaché au rang d'un chiffre dans un système de numérotation.  
- **bit de plus faible poids** : le bit situé le plus à droite dans un mot : Least Significant Byte.
- **bit de plus fort poids** : le bit situé le plus à gauche dans un mot : Most Significant Byte.

## Exemple : binaire => hexadécimal => décimal
Prenons le binaire : `01011101`.  

**Quelle est l'écriture hexadécimale ?**  
On calcule chaque groupe de 4 bits dans le système décimal :
* `0101` = 2<sup>2</sup> + 2<sup>0</sup> = 5 = `5` (en hexadécimal)
* `1101` = 2<sup>3</sup> + 2<sup>2</sup> + 2<sup>0</sup> = 13 = `D` (en hexadécimal)

Résultat hexadécimal = `5D`  
Conclusion avec les préfixes : `0b01011101` = `0x5D`

**Et quelle est cette valeur dans le système décimal ?**  
Il faut calculer `5D` en base 16 :  
`5D` = 5 x 16<sup>1</sup> + 13 x 16<sup>0</sup> = 5 x 16 + 13 = 93

Conclusion : `0b01011101` = `0x5D` = 93

```python
#### Vérification
# Pour convertir binaire => hexadécimal, il faut passer par les int.

# 1/ convertir : binaire => int
entier = int('01011101', 2)

# 2/ Puis : int => hexadécimal
chaine_hexa = hex(entier)

print(chaine_hexa)
print(entier)
```

# Octal = base 8
## Un symbole octal = 3 bits
De manière analogue, l’idée est de regrouper les bits 3 par 3.  
Et chaque bloc de 3 bits par un symbole : **0, 1, 2, ..., 7**

Petit rappel : un mot de 3 bits permet de coder 8 entiers allant de .... à .....

Le préfixe Python d'une écriture octale est `0o`

---
*Exemple* : Prenons le binaire : `01 011 101` :
* `01` = `001` = 1
* `011` = 2<sup>1</sup> + 2<sup>0</sup> = 3
* `101` = 2<sup>2</sup> + 2<sup>0</sup> = 5

Finalement, `01 011 101` (binaire) = `135`(octal)  

<span class="enonce">Convertir `135`(octal) dans le système décimal :</span> ` ...` x 8<sup>2</sup> + ` ....` x 8<sup>1</sup> + `...` x 8<sup>0</sup> = 93

**Conclusion** : En notant la base en indice : `01 011 101`<sub>2</sub> = `135`<sub>8</sub> = 93<sub>10</sub>

```python
### décimal => octal
print("L'entier 93 en octal : ",oct(93))

### octal => décimal
print("L'octal 135 en décimal : ",int("135",8))

### décimal => octal (via les int)
n = int("01011101",2)
ch_octale = oct(n)
print("Le binaire 01011101 en octal :",ch_octale)
```

## Les droits linux : rwx
Cette écriture octale est utilisée dans l’architecture Linux où il existe 3 types de droit : 
> **rwx** : lecture/écriture/exécution (read/write/execute).

Puis, chaque fichier/dossier possède un utilisateur propriétaire et un groupe propriétaire (le groupe pouvant être constitué d’un ou plusieurs utilisateurs).

Et des droits **rwx** sont attribués :
<br>• U : à l’<span style="font-weight:bold;">U</span>tilisateur 
<br>• G : au <span style="font-weight:bold;">G</span>roupe 
<br>• O : aux autres : <span style="font-weight:bold;">O</span>ther


*__Exemple__ :* <span style="background-color:#CCD1D1;">&nbsp;-&nbsp;</span><span style="background-color:#2ECC71;">&nbsp;r w -&nbsp;</span><span style="background-color:#F4D03F;">&nbsp;r - -&nbsp;</span><span style="background-color:#E74C3C;">&nbsp;r - -&nbsp;</span> &nbsp; signifie :
* <span style="background-color:#CCD1D1;">&nbsp;-&nbsp;</span> : ce n'est pas un dossier, donc c'est un fichier.
* <span style="background-color:#2ECC71;">&nbsp;r w -&nbsp;</span> : l'utilisateur propriétaire a le droit de lecture/écriture (read/write).
* <span style="background-color:#F4D03F;">&nbsp;r - -&nbsp;</span><span style="background-color:#E74C3C;">&nbsp;r - -&nbsp;</span> : le groupe propriétaire et les autres ont le droit de lecture (read).
